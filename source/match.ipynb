```python
sido_l = glob.glob("../data/result/*_전체.xlsx")
```

```python
for d in sido_l:
    sido = d[-10:-8]
    a7 = pd.DataFrame()
    for k, j in enumerate(b):

        a6 = pd.DataFrame(j.loc[j["시도"] == sido].groupby("구").count().reset_index()["구"])

        for i in range(12):
            a1 = j.loc[(j["일시"].dt.month == i+1) & (j["시도"] == sido)]
            a2 = a1.iloc[:,[1,2,3,5]]
            a3 = round(a2.groupby("구")[["기온(°C)","풍속(m/s)"]].mean(),1).reset_index()
            a4 = pd.DataFrame(a2.groupby("구")["강수량(mm)"].sum()).reset_index()
            a5 = pd.merge(a3,a4, on="구")
            a5.columns=["구",f"{i+1}월_평균기온(°C)",f"{i+1}월_평균풍속(m/s)",f"{i+1}월_총_강수량(mm)"]
            a6 = pd.merge(a6,a5, on="구")
        a6["연도"] = k+2018

        a7 = pd.concat([a7,a6])

    b1 = pd.read_excel(f"../data/result/{sido}_전체.xlsx")

    b2 = pd.merge(b1,a7, on = ["연도", "구"], how = "left")
    b2.to_excel(f"../data/result/awsd/{sido}.xlsx", index=False)
```

```python
f_l = glob.glob("../data/aws/empty/*.csv")
for i in f_l:
    d_year = date_year.copy()

    for j in range(12):
        tmp = pd.read_csv(i, encoding="cp949")
        tmp = tmp.iloc[:, 2:]
        tmp["일시"] = pd.to_datetime(tmp["일시"] , format="%Y-%m")
        tmp1 = tmp[tmp["일시"].dt.month == j+1].copy()
        tmp1["일시"] = tmp1["일시"].dt.year
        tmp1.columns = ["일시", f"{j+1}월_평균기온(°C)", f"{j+1}월_평균풍속(m/s)", f"{j+1}월_총_강수량(mm)"]
        d_year = pd.merge(d_year, tmp1, on="일시", how = "left")
    
    d_year.to_csv(i, encoding= "utf-8",index=False)
```
