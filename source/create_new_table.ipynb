```python
import pandas as pd
import numpy as np
import glob
```

```python
name_dict = {0:"season", 1:"weather", 2:"law", 3:"acc1", 4:"acc2", 
             5:"time", 6:"age", 7:"day", 8:"career", 9:"sex", 10:"month", 11:"drink", 12:"holiday"}
data = {}

name_l = glob.glob("../result/서울_*.xlsx")
len(name_l)
```

```python
for i,j in enumerate(name_l):
    tmp_df = pd.read_excel(j)
    data[name_dict[i]] = tmp_df.iloc[:,1:]
data
```

```python
tmp_data = pd.DataFrame()
for i in [0,1,2,5,6,7,8,9,10,11,12]:
    tmp = data[name_dict[i]].iloc[:,3:]
    tmp_data = pd.concat([tmp_data ,tmp], axis= 1)
tmp_data
```

```python
data_all = pd.DataFrame()
data_all = pd.concat([data_all,data["sex"].iloc[:,0:3]])
data_all["발생_전체"] = data["sex"]["발생_남성"] + data["sex"]["발생_알수없음"] + data["sex"]["발생_여성"]
```

```python
data_all["사망_전체"] = data["sex"]["사망_남성"] + data["sex"]["사망_알수없음"] + data["sex"]["사망_여성"]
data_all
```

```python
name_l = glob.glob("../data/sido/*")
for i in name_l:
    file_l = glob.glob(f"{i}/*")
    
    for k, j in enumerate(file_l):
        tmp = pd.read_excel(j)
        tmp = tmp.iloc[:,1:]
        if k == 0:
            first = tmp
        first = pd.merge(first,tmp, on = ["시도", "구", "연도"])
    first = first.fillna(0)
    first["발생_전체"] = first.iloc[:,3:7].sum(axis=1)
    first["사망_전체"] = first.iloc[:,7:11].sum(axis=1)
    first.to_excel(f"../data/result/{i[-2:]}_전체.xlsx", index=False)
```

```python
file_l = glob.glob("../data/sido/대구/*")
    
for k, j in enumerate(file_l):
    tmp = pd.read_excel(j)
    tmp = tmp.iloc[:,1:]
    if k == 0:
        first = tmp
    first = pd.merge(first,tmp, on = ["시도", "구", "연도"], how = "left")
first = first.fillna(0)
first["발생_전체"] = first.iloc[:,3:7].sum(axis=1)
first["사망_전체"] = first.iloc[:,7:11].sum(axis=1)
first.to_excel(f"../data/result/대구_전체.xlsx", index=False)
```
