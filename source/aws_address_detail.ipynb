```python
def weather_detail():
    #류가연이 담당한 파트 : exception_l[8]+1,exception_l[12]
        exception_l = [2, 85, 147, 191, 252, 259, 266, 271, 287, 317, 323, 332, 355, 434, 470, 511, 546]
        check_l = range(exception_l[8]+1,exception_l[9])
        cc = [x for x in check_l if x not in exception_l]
        
        driver = webdriver.Chrome(options=chrome_options)
        driver.get("https://data.kma.go.kr/data/grnd/selectAwsRltmList.do?pgmNo=56")
        time.sleep(1)

        for exp in exception_l:
            plus = driver.find_element(By.ID,f"ztree_{exp}_switch")
            plus.click()
            time.sleep(1)
        
        #여기서 loc_check는 화면에 보이게 하고싶은 웹페이지 요소
        loc_check = driver.find_element(By.ID,"ztree_3_check")
        #loc_check가 화면상에 보이게끔 자동으로 스크롤해줌
        driver.execute_script("arguments[0].scrollIntoView();", loc_check)
        
        try:
            dic = {}
            for exp in check_l:
                if exp in [62, 95, 98, 99, 102, 110, 116, 127, 133, 144, 180, 303, 418, 421, 437, 453, 529]:  # 요소의 내용이 비어 있는지 확인
                    continue  # 내용이 비어 있다면 반복문 다음으로 넘어감
                tmp = driver.find_element(By.XPATH,f'//*[@id="ztree_{exp}_a"]/label').text
                driver.find_element(By.XPATH, f'//*[@id="ztree_{exp}"]/a[4]').click()
                time.sleep(3)
                check =  driver.find_element(By.CSS_SELECTOR, \
                '#wrap-datapop > div > div.cont_layer.info_point > div:nth-child(2) > table > tbody > tr:nth-child(4) > td:nth-child(4)').text
                time.sleep(3)
                dic[tmp] = check.split()[:2]
                print(tmp , dic[tmp])
                driver.find_element(By.XPATH, '//*[@id="wrap-datapop"]/div/div[1]/button').click()
                time.sleep(3)
        except NoSuchElementException:
            print("요소를 찾을 수 없습니다.")
            
        return dic
```
